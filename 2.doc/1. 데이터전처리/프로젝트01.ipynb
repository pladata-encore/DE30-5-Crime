```python
import pandas as pd
```

```python
df = pd.read_csv('police.csv',encoding='cp949')
df
```

```python
test = pd.read_csv('police.csv',encoding='cp949') #데이터 받아서 인코딩
test1 = test.drop(columns=['범죄중분류'],axis=0) # 범죄 중분류 행 없애기
test2 = test1.groupby(['범죄대분류']).sum() #그룹바이를 이용해 대분류끼리 합 구하기
df_filtered = test2[[column for column in test2.columns if '경기' in column]]
b1 = df_filtered.T.sum() #경기라는 이름을 가진 컬럼 찾기
df_filtered = test2[[column for column in test2.columns if '강원' in column]]
b2 = df_filtered.T.sum()
df_filtered = test2[[column for column in test2.columns if '충북' in column]]
b3 = df_filtered.T.sum()
df_filtered = test2[[column for column in test2.columns if '충남' in column]]
b4 = df_filtered.T.sum()
b31 = b3+b4
df_filtered = test2[[column for column in test2.columns if '전북' in column]]
b5 = df_filtered.T.sum()
df_filtered = test2[[column for column in test2.columns if '전남' in column]]
b6 = df_filtered.T.sum()
b41=b5+b6
df_filtered = test2[[column for column in test2.columns if '경북' in column]]
b7 = df_filtered.T.sum()
df_filtered = test2[[column for column in test2.columns if '경남' in column]]
b8 = df_filtered.T.sum()
b51=b7+b8
df_filtered = test2[[column for column in test2.columns if '제주' in column]]
b9 = df_filtered.T.sum()
merge_data = pd.concat([b1,b2,b31,b41,b51,b9],axis=1) #'경기도','강원도','충청도','전라도','경상도','제주도' 합치기
merge_data.columns=['경기도','강원도','충청도','전라도','경상도','제주도'] #컬럼 이름 붙여주기
merge_data
a1 = (test2).iloc[:,0:8]  #대도시
a2 = merge_data #도단위 합친거 
a3 = (test2).iloc[:,-2:]  #기타 및 도시이외
result = pd.merge(a1, a2, on='범죄대분류', how='inner') #도단위와 대도시 합치기
realrealdata = pd.merge(result, a3, on='범죄대분류', how='inner')#도단위와 대도시 합친거에 기타도 합치기
realrealdata
```

```python
realrealdata.to_csv('police02.csv',encoding='utf-8',header=False)
df = pd.read_csv('police02.csv',header=None)
df
```
